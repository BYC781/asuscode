### 三立

```python
URL='https://www.setn.com/search.aspx?kw=ASUS&r=0'
```

```python
google.get(URL)
```

```python
google.find_element(By.XPATH, '//*[@id="keyword"]').send_keys('ASUS')
google.find_element(By.XPATH, '//*[@id="btnSearch"]/i').click()
```

```python
results = google.find_elements(By.XPATH, '//*[@id="contFix"]/div/div[2]/div[4]/div/div')

# 初始化空的DataFrame
news_info = pd.DataFrame(columns=["Title", "Date", "URL"])

# 遍歷每個元素，並獲取其標題、日期和URL
for result in results:
    title = result.find_element(By.XPATH, './a/div[3]').text
    date = result.find_element(By.XPATH, './div[2]/div[1]').text
    url = result.get_attribute("href")
    row = {"Title": title, "Date": date, "URL": url}
    news_info = pd.concat([news_info, pd.DataFrame([row])], ignore_index=True)
```

### 東森

```python
import pandas as pd
import time
ebc_news_info = pd.DataFrame(columns=['Title', 'Date', 'URL', 'Content'])
ebc_news_info = ebc_news_info.astype({'Title': str, 'Date': str, 'URL': str, 'Content': str})
```

```python
google.get('https://news.ebc.net.tw/')
google.get('https://news.ebc.net.tw/search/result/keyword?value=ASUS')
time.sleep(7)
```

```python
results = google.find_elements(By.XPATH, '//*[@id="contentb"]/div[1]/div[2]/div/a')
for result in results:
    title = result.find_element(By.XPATH, './div[2]/span[1]').text
    date = result.find_element(By.XPATH, './div[2]/span[3]').text
    href = result.get_attribute('href')
    row = {"Title": title, "Date": date, "URL": href}
    ebc_news_info = pd.concat([ebc_news_info, pd.DataFrame([row])], ignore_index=True)
```

```python
cnt = 1
for idx in range(ebc_news_info.shape[0]):
    google.get(ebc_news_info.loc[idx, 'URL'])
    content = google.find_elements(By.XPATH, '//*[@id="main_content"]/div[5]/div/p')
    con_str = []
    for c in content:
        con_str.append(c.text)
    ebc_news_info.loc[idx, 'content'] = ["\n".join(con_str)]
    if cnt == 100:
        break
    cnt += 1
```

```python
google.find_element(By.XPATH, '//*[@id="search"]').click()
google.find_element(By.XPATH, '//*[@id="search"]').send_keys('ASUS')
google.find_element(By.XPATH, '//*[@id="search"]').send_keys(Keys.ENTER)
```

```python
# 排序依據改成日期
google.find_element(By.XPATH, '//*[@id="___gcse_0"]/div/div/div[1]/div[4]/table/tbody/tr/td[2]/div/div[2]/div[1]').click()
google.find_element(By.XPATH, '//*[@id="___gcse_0"]/div/div/div[1]/div[4]/table/tbody/tr/td[2]/div/div[2]/div[2]/div[2]').click()
```

```python
title_and_href = google.find_elements(By.XPATH, '//*[@id="___gcse_0"]/div/div/div[1]/div[6]/div[2]/div/div/div[1]/div/div[1]')
date_pattern = r"\d{4}年\d{1,2}月\d{1,2}日"
for ele in title_and_href:
    title = ele.find_element(By.XPATH, './div[1]/div/a').text
    date = re.findall(date_pattern, ele.find_element(By.XPATH, './div[3]/div/div[2]').text)[0]
    href = ele.find_element(By.XPATH, './div[1]/div/a').get_attribute('href')
    row = {"Title": title, "Date": date, "URL": href, 'Content': ''}
    tvbs_news_info = pd.concat([tvbs_news_info, pd.DataFrame([row])], ignore_index=True)
    
```
