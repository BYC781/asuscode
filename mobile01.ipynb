```python
from selenium.common.exceptions import NoSuchElementException
import os
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
import re
from selenium.webdriver.common.keys import Keys
import time
```

```python
os.system(r'start chrome --remote-debugging-port=9527 --user-data-dir="c:\Users\ElvisBY_Chen\OneDrive - ASUS\Documents\Python Scripts\news"')
options = Options()
options.add_experimental_option("debuggerAddress", "127.0.0.1:9527")
google = webdriver.Chrome(options=options)
```

```python
mobile01_info = pd.DataFrame(columns=['Title', 'URL', 'Catagory', 'Date'])
```

```python
google.get('https://www.mobile01.com/')
google.find_element(By.XPATH, '//*[@id="keyword"]').click()
google.find_element(By.XPATH, '//*[@id="keyword"]').send_keys('ASUS')
google.find_element(By.XPATH, '//*[@id="keyword"]').send_keys(Keys.ENTER)
```

```python
date_pattern = r"\d{4}年\d{1,2}月\d{1,2}日"
results = google.find_elements(By.XPATH, '//*[@id="___gcse_0"]/div/div/div/div[5]/div[2]/div/div/div[1]/div')
cnt = 0
for res in results:
    cnt += 1
    print(cnt)
    if res.get_attribute('textContent') != '':
        url = res.find_element(By.XPATH, './div[1]/div[1]/div/a').get_attribute('href')
        title = res.find_element(By.XPATH, './div[1]/div[1]/div/a').text
        catagory = res.find_element(By.XPATH, './div[1]/div[2]/div[3]').get_attribute('textContent')
        try:
            date = re.findall(date_pattern, res.find_element(By.XPATH, './div[1]/div[3]/div[2]/div[2]').text)[0]
        except NoSuchElementException:
            date = re.findall(date_pattern, res.find_element(By.XPATH, './div[1]/div[3]/div/div[2]').text)[0]
        row = {"Title": title, "Date": date, "URL": url, 'Catagory': catagory}
        mobile01_info = pd.concat([mobile01_info, pd.DataFrame([row])], ignore_index=True)
```
