```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from random import randint
import pandas as pd
```

```python
def make_df():   
    brand = []
    title = []
    url = []
    price = []
    star = []
    starNum = []
    feature = []
    productDscrp = []
    view_url = []
    dic = {
        'brand' : brand,
        'title' : title,
        'url' : url,
        'price' : price,
        'star' : star,
        'starNum' : starNum,
        'feature' : feature,
        'productDscrp' : productDscrp,
        'view_url' : view_url
        }
    df = pd.DataFrame(dic)

    data_types = {
        'brand': str,
        'title': str,
        'url': str,
        'price': str,
        'star': str,
        'starNum': str,
        'feature': str,
        'productDscrp': str,
        'view_url': str
    }
    df = df.astype(data_types)
    return df
```

```python
def process(firm):
  df = make_df()
  for page in range(start, len(theurl)):
      print('第 '+ str(page) + ' 個商品')
      #儲存網址
      df.loc[page, 'url']= theurl[page]
      
      # 去到你想要的網頁
      driver.get(theurl[page])
      time.sleep(randint(2, 5))

      # 品牌名稱
      if len(driver.find_elements(By.ID, 'bylineInfo')) == 0 :
          df.loc[page, 'brand'] = 'No brand'
      else:
          df.loc[page, 'brand'] = driver.find_element(By.ID, 'bylineInfo').text

      # 商品名稱
      df.loc[page, 'title'] = driver.find_element(By.ID, 'title').text

      # 價格
      if len(driver.find_elements(By.XPATH, '//*[@id="corePrice_feature_div"]')) != 0:
          getprice = driver.find_element(By.XPATH, '//*[@id="corePrice_feature_div"]').text
          getprice = getprice.replace('US$', '').replace('\n', '.')
          try:
              getprice = float(getprice)
          except ValueError:
              getprice = ''
          df.loc[page, 'price'] = getprice
      else:
          df.loc[page, 'price'] = ''
      
      # 星星評分
      if len(driver.find_elements(By.ID, 'acrPopover'))==0:
          df.loc[page, 'star'] ='No star'
      else:
          df.loc[page, 'star'] =driver.find_element(By.ID, 'acrPopover').get_attribute("title").replace(' 顆星，最高 5 顆星','')

      # 星星數量
      if len(driver.find_elements(By.ID, 'acrCustomerReviewText')) == 0:
          df.loc[page, 'starNum'] = 0
      else:
          getglobalNum = driver.find_element(By.ID, 'acrCustomerReviewText').text
          getglobalNum = getglobalNum.replace(' 個評分','').replace(',','')
          df.loc[page, 'starNum'] = getglobalNum

      # product feature
      if len(driver.find_elements(By.ID, 'productOverview_feature_div')) != 0:
          df.loc[page, 'feature'] = driver.find_element(By.ID, 'productOverview_feature_div').text
      else:
          df.loc[page, 'feature'] = ''

      # prod descrip
      if len(driver.find_elements(By.ID, 'productDescription')) != 0:
          df.loc[page, 'productDscrp'] = driver.find_element(By.ID, 'productDescription').text
      else:
          df.loc[page, 'productDscrp'] = ''


      # 留言網址
      if len(driver.find_elements(By.XPATH,'//a[@data-hook = "see-all-reviews-link-foot"]'))== 0 :
          df.loc[page, 'view_url'] = 'No review'
      else:
          df.loc[page, 'view_url'] = driver.find_element(By.XPATH,'//a[@data-hook = "see-all-reviews-link-foot"]').get_attribute('href')
  
  return df

  
```
